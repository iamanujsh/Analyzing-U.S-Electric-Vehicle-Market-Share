```python
import pandas as pd
import numpy as np
```

```python
# Importing and removing the thouands seperator

df = pd.read_csv('/content/Vehicle Data raw.csv', thousands= ',')
```

```python
# Sortaning / Renaming Column Names - Standaried State Names

df = df.rename(columns= {'Electric (EV)' : "EV" , 'Plug-In Hybrid Electric (PHEV)'  : 'PHEV' ,
                                  'Hybrid Electric (HEV)' : 'HEV' , 'Ethanol/Flex (E85)' : 'E85' ,
                                  'Compressed Natural Gas (CNG)' : 'CNG'
                                 })
```

```python
# Calculate the percentage of EVs, PHEVs, HEVs, and Gasoline vehicles for each state.

df['Total Vehicles'] = (df['EV'] + df['PHEV'] + df['HEV'] + df['Biodiesel'] + df['E85']  +
                        df['CNG'] + df['Propane'] + df['Hydrogen'] + df['Methanol'] + df['Gasoline'] + df['Diesel'] + df['Unknown Fuel'])

df['% of EV'] = df['EV']  / df['Total Vehicles'] * 100
df['% of PHEVs'] = df['PHEV']  / df['Total Vehicles'] * 100
df['% of HEVs'] = df['HEV']  / df['Total Vehicles'] * 100
df['% of Gasoline'] = df['Gasoline']  / df['Total Vehicles'] * 100
```

```python
# Identify the top 5 states with the highest EV adoption rate (EVs as a % of all registered vehicles).

df[['State', 'Total Vehicles' ,'% of EV']].sort_values('% of EV' , ascending = False).head()
```

```python
# Compare EV adoption in California vs. other large states (e.g., Texas, Florida, New York).

condition = df['State'].isin(['California','Texas' ,'Florida', 'New York'])

df[['State', 'Total Vehicles' ,'% of EV']].loc[condition].sort_values('% of EV', ascending = False)
```

```python
# Highlight which alternative fuels (biodiesel, ethanol, hydrogen) have meaningful presence vs. niche usage.

target_fuels = ['Biodiesel', 'E85', 'Hydrogen']
total_target_fules = df[target_fuels].sum()
total_vehicles = df['Total Vehicles'].sum()

percent = (total_target_fules / total_vehicles) * 100

niche_vs_broad = pd.DataFrame({
    'Total Fules' : total_target_fules,
    'percent' : percent
})

niche_vs_broad['presence'] = niche_vs_broad['percent'].apply(lambda x : "Meaningful" if x > 1.0 else "Niche")

niche_vs_broad
```
